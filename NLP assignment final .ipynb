```python
import pandas as pd 
import numpy as np
import nltk 
import PyPDF2 as pdf
import re 
import os 
```

```python
path_h = 'C:/Users/Ishant Naru/Desktop/NLP Assignment/History/'

list_h = os.listdir(path_h)
#list_s = os.listdir(path_s)
history = []
for i in list_h:
    file_path = os.path.join(path_h + i)
    file = pdf.PdfFileReader(file_path)
    history.append(file)

# science = []
# for i in list_s:
#     file_path = os.path.join(path_s + i)
#     file = pdf.PdfFileReader(file_path)
#     science.append(file)

history
```

```python
#Extracting text form all files into a list of strings 
history_list = []
for ele in history:
    for i in range(ele.getNumPages()):
        string = ''
        string = string + ele.getPage(i).extractText()
        history_list.append(string)

# science_list = []
# for ele in science:
#     for i in range(ele.getNumPages()):
#         string = ''
#         string = string + ele.getPage(i).extractText()
#         science_list.append(string)
```

```python
history_list[7]
```

```python
cl_history_list = []
for string in history_list:
    string = re.sub('\n', '', string)
    cl_history_list.append(string)

cl_history_list
```

```python
#removing empty string elements from the list
while ('' in cl_history_list):
    cl_history_list.remove('')

cl_history_list
```

```python
from nltk.stem import WordNetLemmatizer as WNL
from nltk.corpus import stopwords 

Lemm = WNL()
```

```python
#converting all elements of list into a single string for preprocessing 
history_string = ','.join(cl_history_list)
history_string
```

```python
#tokenize
sentences = nltk.sent_tokenize(history_string)

sentences
```

```python
#Applying stopwords, cleaning and lower casing the text 
corpus_h = []
for i in range(len(sentences)):
    cleaned = re.sub('[^a-zA-Z]', ' ', sentences[i])
    cleaned = cleaned.lower()
    cleaned = cleaned.split()
    
    words = [Lemm.lemmatize(word) for word in cleaned if not word in set(stopwords.words('english'))]
    words = ' '.join(words)
    corpus_h.append(words)
    
corpus_h
```

```python
#removing empty string elements from the list
while ('' in corpus_h):
    corpus_h.remove('')

corpus_h
```

```python
#Applying TF-IDF Vectorizer to history questions words and converting them into a dataframe 
from sklearn.feature_extraction.text import TfidfVectorizer as TIV

TIV = TIV()
X_h = TIV.fit_transform(corpus_h).toarray()

Hist_df = pd.DataFrame(X_h)
Hist_df['label'] = 'history'
Hist_df.head(), Hist_df.shape
```

```python
#converting the list of questions into a single string 
science = ','.join(science_list)
science_string = re.sub('\n', ' ', science)
science_string
```

```python
#Tokenize science ques
sentences_s = nltk.sent_tokenize(science_string)
sentences_s
```

```python
#Applying stopwords, cleaning and lower casing the text 
corpus_s = []
for i in range(len(sentences_s)):
    cleaned = re.sub('[^a-zA-Z]', ' ', sentences_s[i])
    cleaned = cleaned.lower()
    cleaned = cleaned.split()
    
    words = [Lemm.lemmatize(word) for word in cleaned if not word in set(stopwords.words('english'))]
    words = ' '.join(words)
    corpus_s.append(words)
    
corpus_s
```

```python
#removing empty string elements from the list
while ('' in corpus_s):
    corpus_s.remove('')

corpus_s
```

```python
#Applying TF-IDF vectorizer technique and saving Science Dataframe
X_s = TIV.fit_transform(corpus_s).toarray()

Sci_df = pd.DataFrame(X_s)
Sci_df['label'] = 'science'
Sci_df.head(), Sci_df.shape
```

```python
#Combining the dataframes of history and science 
dfs = [Hist_df, Sci_df]
dataset = pd.concat(dfs)
dataset.tail()

dataset.to_csv('Final_Dataset_df.csv')
```

```python
#Creating Independent and Dependent Variables and One-hot encoding the dependent variable 
X = dataset.drop(labels = ['label'], axis=1)
y = dataset['label']
y = pd.get_dummies(y)
y = y.iloc[:,1]
```

```python
#Train Test Split 
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=42)

X_train.shape, X_test.shape
```

```python
from sklearn.impute import SimpleImputer 

impute = SimpleImputer(missing_values = np.nan, strategy = 'constant')
X_train = impute.fit_transform(X_train)
X_test = impute.transform(X_test)

X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```

```python
#Fitting the Naive Bayes model 
from sklearn.naive_bayes import MultinomialNB as MNB

model = MNB().fit(X_train, y_train)

#Predicting
y_pred = model.predict(X_test)
```

```python
#Evaluation of Performance for Naive Bayes Model 
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_fscore_support as PRF_summary
CM = confusion_matrix(y_test, y_pred)

CM 

results = classification_report(y_test, y_pred)

precision, recall, f1, support = PRF_summary(y_test, y_pred)
print('Precision for hist and science resp: ',precision)
print('Recall for hist and science resp: ',recall)
print('f1 score for hist and science resp: ',f1)
```

```python
#Logistic Regression Model 
from sklearn.linear_model import LogisticRegression as LR 

classifier = LR(class_weight='balanced')
LR_model = classifier.fit(X_train, y_train)
LR_pred = LR_model.predict(X_test)
```

```python
#Evaluation of Performance for Logistic Regression Model 
precision_LR, recall_LR, f1_LR, support_LR = PRF_summary(y_test, LR_pred)
print('Precision for hist and science resp: ',precision_LR)
print('Recall for hist and science resp: ',recall_LR)
print('f1 score for hist and science resp: ',f1_LR)
```
